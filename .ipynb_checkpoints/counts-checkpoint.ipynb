```python
import pandas as pd 
import datetime
```

```python
for k in range(1,51):
    base_str = 'C:\\Users\\raosa\\Dropbox\\data\\combined\\'
    data = pd.read_csv('C:\\Users\\raosa\\Dropbox\\dl-graph\\data\\eth\\pricing\\final.csv')
    x=pd.read_csv(str(base_str+str(k)+'.csv'))
    print('reading done')
    print(len(x))
    timestamp_list = x['timestamp']
    value_list = x['value']

    timestamp_list_1=[]
    value_list_1=[]
    print('loop starting')
    for i in  range(0,len(timestamp_list)):
        test1 = datetime.datetime.fromtimestamp(int(str(timestamp_list[i]),16))
        test2 = (int(x['value'][i],16))/1000000000000000000
        timestamp_list_1.append(test1)
        value_list_1.append(test2)
    print('loop done')
    x['timestamp1'] = timestamp_list_1
    x['value1'] = value_list_1


    x=x.drop(['timestamp','value'],axis=1)
    x = x.rename(columns={'timestamp1': 'timestamp', 'value1': 'value'})



    #     value_list = x['value']
    #     for i in range(0,len(x)): 
    #         x['timestamp1'] = datetime.datetime.fromtimestamp(int(str(x['timestamp'][i]),16))
    #         x['value1'][i] = int(x['value'][i],16)

    #     print('for loop done')
    x['timestamp'] = pd.to_datetime(x['timestamp'])
    # group by the number of transactions per hour 
    tran_counts = x.groupby([pd.Grouper(key='timestamp',freq='H')]).size().reset_index(name='count')
    print('groupby done')
    # convert to datetime object 
    data['Date'] = pd.to_datetime(data['Date'])
    # merge both datasets 
    y=pd.merge(tran_counts,data,left_on='timestamp',right_on='Date')
    y=y.drop(['timestamp','pct_change'],axis=1)
    y.to_csv(str('count_'+str(k)+'.csv'),index=False)
    print(k)
    print('_______________________')
```
